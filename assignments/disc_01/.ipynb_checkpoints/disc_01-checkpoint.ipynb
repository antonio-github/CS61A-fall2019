## Q1.

```python
def wears_jacket_with_if(temp, raining):
    """
    >>> wears_jacket_with_if(90, False)
    False
    >>> wears_jacket_with_if(40, False)
    True
    >>> wears_jacket_with_if(100, True)
    True
    """
    if temp < 60:
        return True
    elif raining == True:
        return True
    else:
        return False
```

```python
def wears_jacket(temp, raining):
    """
    >>> wears_jacket(90, False)
    False
    >>> wears_jacket(40, False)
    True
    >>> wears_jacket(100, True)
    True
    """
    return (temp < 60) or (raining == True)
```

In the solutions `(raining == True)` is simplified to `raining` because it already evaluates to `True` or `False`.


## Q1.2.

```python
def square(x):
    print("here!")
    return x * x

def so_slow(num):
    x = num
    while x > 0:
        x = x + 1
    return x / 0

square(so_slow(5))
```

`so_slow(5)` never returns any value because `x > 0` is always true. Therefore there is no result in evaluating that `square(so_slow(5))`.


## Q.1.3

```python
def is_prime(n):
    """
    >>> is_prime(10)
    False
    >>> is_prime(7)
    True
    """
    divisor = 2
    while divisor < n:
        if (n%divisor) == 0:
            return True
        divisor += 1
    return False
```
